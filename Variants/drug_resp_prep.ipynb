```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load both files

dose_response_df = pd.read_csv('../data/v17.3_fitted_dose_response.csv', index_col='IC50_RESULTS_ID')
dose_response_df.head()

wes_scored = pd.read_csv("../data/wes_scored_transposed.tsv",sep = "\t")
#wes_scored = pd.read_csv("../data/wes_scored_non_binary_transposed.tsv",sep = "\t", index_col='COSMIC_ID')
wes_scored.set_index("Unnamed: 0",drop= True ,inplace=True)
wes_scored.index.names = ['COSMIC_ID']
wes_scored.head()
```

```python
cna_scored = pd.read_csv("../data/cna_scored_transposed.tsv",sep = "\t", index_col='COSMIC_ID')
cna_scored.head()
```

```python

def generate_wes_response(drug_id, drug_name):
    
    dose_by_drug_df = dose_response_df[dose_response_df.DRUG_ID == drug_id]
    dose_response_trimmed_df = dose_by_drug_df[['COSMIC_ID', 'DRUG_ID', 'LN_IC50']]
    
    joined_df = pd.merge(wes_scored, dose_response_trimmed_df, on='COSMIC_ID')
    file_name = "../data/wes_ic50_"+ drug_name + ".tsv"
    joined_df.to_csv(file_name,sep = "\t")
    
    
```

```python
def generate_cna_response(drug_id, drug_name):
    
    dose_by_drug_df = dose_response_df[dose_response_df.DRUG_ID == drug_id]
    dose_response_trimmed_df = dose_by_drug_df[['COSMIC_ID', 'DRUG_ID', 'LN_IC50']]
    
    joined_df = pd.merge(cna_scored, dose_response_trimmed_df, on='COSMIC_ID')
    file_name = "../data/cna_ic50_"+ drug_name + ".tsv"
    joined_df.to_csv(file_name,sep = "\t")
```

```python
drug_df = pd.read_csv('../data/Screened_Compounds.csv', index_col='DRUG_ID')
```

```python
for index, row in drug_df.head(10).iterrows():
    generate_wes_response(index, row["DRUG_NAME"])
    #generate_cna_response(index, row["DRUG_NAME"])
```
